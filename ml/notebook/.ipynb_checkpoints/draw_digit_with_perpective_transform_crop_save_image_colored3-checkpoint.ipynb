```python
image_transform = cv2.imread("data3/Img_10_473386144.bmp")
digits_before = 4
digit_after = 2

# Before dp 
for digit in range(digits_before):
    if digit >= 0: 
        if digit == 3 or digit ==4: 
            continue 
        initial_x = 47
        x = initial_x + digit*15 
        y = 15
        w = 15
        h = 25
        print(digit)
        cropped = cv2.rectangle(image_transform,(x,y),(x+w,y+h),(0,255,0),1)
        plt.imshow(cropped)
        plt.show()

# After dp
for digit in range(digit_after):
    if digit >= 0:
        initial_x = 89
        x = initial_x + digit*15 
        y = 15
        w = 15
        h = 25
        print(digit)
        cropped = cv2.rectangle(image_transform,(x,y),(x+w,y+h),(0,255,0),1)
        plt.imshow(cropped)
        plt.show()
```

## Crop and Save images

```python
import os 
import uuid 
```

```python
path = "data3/"
```

```python
digits_before = 4
digit_after = 2

for file in os.listdir(path): 
    image_transform = cv2.imread("data3/" +file)
    print(image_transform)

    # Before dp 
    for digit in range(digits_before):
        if digit >= 0: 
            if digit == 3 or digit ==4: 
                continue 
            initial_x = 46
            x = initial_x + digit*15 
            y = 15
            w = 15
            h = 25
            gen = str(uuid.uuid4())
            print(digit)
            cropped = image_transform
            cropped = cropped[y:y+h, x:x+w]
            cv2.imwrite(f'finished_colored1/{digit}_{gen}.png',cropped)
            plt.imshow(cropped)
            plt.show()

    # After dp
    for digit in range(digit_after):
        if digit >= 0:
            initial_x = 89
            x = initial_x + digit*15 
            y = 15
            w = 15
            h = 25
            gen = str(uuid.uuid4())
            print(digit)
            cropped = image_transform 
            cropped = cropped[y:y+h, x:x+w]
            cv2.imwrite(f'finished_colored1/{digit}_{gen}.bmp',cropped)
            plt.imshow(cropped)
            plt.show()
```
